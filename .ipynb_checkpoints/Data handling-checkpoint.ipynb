## Exercise 1: Processing Student Data with Python

```python
import csv

def process_students_file(input_file, output_file):
    # Read the CSV file
    with open(input_file, mode='r', newline='') as file:
        reader = csv.DictReader(file)
        
        # Filter Data
        filtered_students = []
        for row in reader:
            if row['Grade'] in ['A', 'B']:  # Check if the grade is A or B
                row['Status'] = 'Passed'  # Add Status column
                filtered_students.append(row)

    # Write to a new CSV file
    if filtered_students:
        with open(output_file, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=filtered_students[0].keys())
            writer.writeheader()
            writer.writerows(filtered_students)

process_students_file('students.csv', 'top_students.csv')
```

## Practice Exercise: Extracting State Population

```python
import requests

# Function to get the state data from the API
def get_state_data():
    api_url = "https://jian.sh/malaysia-api/state/v1/all.json"
    response = requests.get(api_url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Failed to retrieve data")
        return None


def get_state_data_by_name(state_name, data):
    for state in data:
        if state['state'].lower() == state_name.lower():
            return state
    return None

def get_by_input_state():
    states_data = get_state_data()
    if states_data:
        # Ask the user for the state name
        user_state = input("Enter the name of a state in Malaysia: ")

        # Find the state data
        state_data = get_state_data_by_name(user_state, states_data)  # Assuming states_data is a list
        if state_data:
            # Now you can print any information you need from state_data
            print(f"State: {state_data['state']}")
            print(f"Capital: {state_data['capital']}")
            print(f"Population: {state_data['population']}")
            # Add any additional data you want to print here
        else:
            print("State not found.")
```
